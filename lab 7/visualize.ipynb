```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("D:\Karthikyan\SECRET\Python\lab 7\IOT-temp.csv")

# Convert 'noted_date' from string to datetime if needed
df['noted_date'] = pd.to_datetime(df['noted_date'], format='%d-%m-%Y %H:%M')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['noted_date'], df['temp'], alpha=0.6)
plt.title('Temperature over Time')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['noted_date'], df['temp'], marker='', linestyle='-')
plt.title('Temperature Trend over Time')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()
```

```python
avg_temp_by_location = df.groupby('out/in')['temp'].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(x='out/in', y='temp', data=avg_temp_by_location)
plt.title('Average Temperature by Location')
plt.xlabel('Location')
plt.ylabel('Average Temperature')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['temp'], bins=30, color='blue', alpha=0.7)
plt.title('Temperature Distribution')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='out/in', y='temp', data=df)
plt.title('Temperature Distribution by Location')
plt.xlabel('Location')
plt.ylabel('Temperature')
plt.show()
```

```python
# Let's assume your dataframe has more numerical columns or you use existing ones
sns.pairplot(df[['temp', 'out/in']])
plt.title('Pair Plot of Data')
plt.show()
```
